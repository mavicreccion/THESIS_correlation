```python
# Import packages
import numpy as np
import pandas as pd
from scipy.stats.stats import spearmanr
```

```python
dataset = pd.read_csv("merged_mma_wwo_manila_2015/merged_mmda_wwo_Taft Ave._2015.csv", skipinitialspace=True, encoding='cp1252')
dataset = dataset[['dt', 'statusN']]
dataset.loc[[1176]]
```

```python
start_index = 1176

data = []

for i in range(1, 31): 
    lag = i

    mask1 = (dataset.index >= start_index) & (dataset.index < start_index + lag)
    data1 = dataset.loc[mask1]
#     print(data1['statusN'].tolist())

    mask2 = (dataset.index >= start_index + lag) & (dataset.index < start_index + lag * 2)
    data2 = dataset.loc[mask2]
#     print(data2['statusN'].tolist())
    corr = spearmanr(data1['statusN'].tolist(), data2['statusN'].tolist()).correlation
    
    row = []
    row.append(lag)
    row.append(corr)
    
    data.append(row)

names = ['lag', 'corr']
corrDF = pd.DataFrame(data, columns=names)
corrDF
```

```python
corrDF.to_csv('traffic_autocorrelation_30lags.csv')
```
