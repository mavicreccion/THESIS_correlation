```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
# Read the dataset
dataset = pd.read_csv("merged_data\\merged_mmda_wwo_manila_2015.csv", skipinitialspace=True, encoding='cp1252')

# Remove unused columms
cols_to_remove = list(range(0,3))
cols_to_remove = cols_to_remove + [6, 7, 9, 17, 19, 21, 22, 25]

dataset.drop(dataset.columns[[cols_to_remove]], axis=1, inplace=True)
dataset.head()
```

```python
# Correlate
corr = dataset.corr(method='spearman')
corr
```
