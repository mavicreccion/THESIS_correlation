```python
# Import packages
import numpy as np
import pandas as pd
```

```python
roads = ["A. Maceda", "Airport Road", "Anda Circle", "Antipolo", "Baclaran", "Bluementritt", "Buendia", "Coastal Road", "Edsa Extension", "Finance Road", "Gov. Forbes - Lacson", "Lerma", "Magsaysay Ave", "P.Noval", "Pablo Ocampo", "Pedro Gil", "Quezon Ave.", "Quirino", "Rajah Sulayman", "Taft Ave.", "U.N. Avenue", "Vicente Cruz", "Welcome Rotunda"]
```

# Merge Data

```python
road_dataset = []
for road in roads:
    dataset = pd.read_csv("merged_mma_wwo_manila_2015/merged_mmda_wwo_" + road + "_2015.csv", skipinitialspace=True, encoding='cp1252')
        
    dataset = dataset[['dt', 'statusS']]
    
    dataset = dataset.set_index(pd.DatetimeIndex(dataset['dt']))
    dataset = dataset.drop(['dt'], axis=1)
    dataset.columns = [road]
    
    road_dataset.append(dataset)
```

```python
traffic_roads = pd.concat(road_dataset, axis=1)
traffic_roads
```

```python
traffic_roads.to_csv('merged_traffic_southbound.csv')
```
